## Driver Standings

```python
import pycountry

nationality_to_country = {
    "New Zealander": "New Zealand",
    "British": "United Kingdom",
    "Dutch": "Netherlands",
    "Monegasque": "Monaco",
    "American": "United States",
}

def get_country_codes(nationality):
    country_name = nationality_to_country.get(nationality, nationality)
    try:
        country = pycountry.countries.search_fuzzy(country_name)[0]
        return country.alpha_2, country.alpha_3
    except LookupError:
        return None, None
```

```python
import pandas as pd

driver_standings = ergast.get_driver_standings(season=2025).content[0]
driver_standings["driver"] = driver_standings["givenName"] + " " + driver_standings["familyName"]
driver_standings_filtered = driver_standings[["position", "driver", "driverCode", "driverNationality", "constructorNames", "constructorIds", "points"]]
driver_standings_filtered[["country_alpha2", "country_alpha3"]] = driver_standings_filtered["driverNationality"].apply(
    lambda x: pd.Series(get_country_codes(x))
)
driver_standings_filtered
```
