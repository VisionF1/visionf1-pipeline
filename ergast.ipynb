# Ergast API

### This file contains tests to find what information is available in the Ergast API through the FastF1 package. This data should be useful for the VisionF1 web app.

Note: the Ergast API was deprecated at the end of 2024. FastF1 now offers the jolpicaf1 API (Ergast compatible) through the same interface.

```python
from fastf1.ergast import Ergast
ergast = Ergast()
```

```python
circuits = ergast.get_circuits(season=2025)
circuits
```

```python
races = ergast.get_race_schedule(season=2025)
races
```
